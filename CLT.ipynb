# Visualizing Central Limit Theorem.


### First we require a population:

```python
# Creating a population of 20,000 people of ages between 20 and 80 both inclusive

import random

population_age= [    random.randint(20,81) for i in range(20001)    ]
```

```python
# Let's take a look at 10 ages here

population_age[:10]
```

```python
# Our goal next is to take a random sample of this population of 10 people which can be done using the random module again.

print(random.sample(population_age, 10))
```

```python
# Let's have a look at the distribution of our population age:

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

sns.distplot(population_age)
plt.show()
```

## We can see that the current distribution is very chaotic

```python


# What CLT says if we take random samples of a population, the mean of those samples turns is approximately normally distributed.

# Let's write a for loop to take 500 random samples of size 10, and store their means into another list
import numpy as np
sample_means= []
count= 0
for i in range(100):
    count += 1
    sample= random.sample(population_age, 10)
    sample_mean= np.mean(sample)
    sample_means.append(sample_mean)
    sns.distplot(sample_means)
    plt.title(f' No. of sample means = {count}')
    plt.show()
   
```

```python
## Now let's have a look at the distribution again

sns.distplot(sample_means)
```

# We can see the distribution changing as the count of sample means increases, we can see it going towards a normal distribution.

```python
# This is the final distribution

sns.distplot(sample_means)
plt.show()

# The shape of final distribution changes everytime because we are taking a different random sample every time.
```
